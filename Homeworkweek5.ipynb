```python
import numpy as np

def getBondPrice(y, face, couponRate, m, ppy=1):

    n = m * ppy                       # total periods
    r = y / ppy                       # yield per period
    C = (couponRate * face) / ppy     # coupon per period

    t = np.arange(1, n + 1)           # time periods
    discount_factors = 1 / (1 + r)**t

    pv_coupons = np.sum(C * discount_factors)
    pv_face = face / (1 + r)**n

    return pv_coupons + pv_face
```

```python
y = 0.03
face = 2000000
couponRate = 0.04
m = 10
ppy = 2

getBondPrice(y, face, couponRate, m, ppy)
```

```python
#geBondDuration
import numpy as np

def getBondDuration(y, face, couponRate, m, ppy=1):

    n = m * ppy
    r = y / ppy
    C = (couponRate * face) / ppy

    t = np.arange(1, n + 1)

    cf = np.full(n, C)
    cf[-1] += face

    discount = 1 / (1 + r)**t
    pvcf = cf * discount

    duration = np.sum(t * pvcf) / np.sum(pvcf)

    return duration
```

```python
y = 0.03
face = 2000000
couponRate = 0.04
m = 10
ppy = 1

getBondDuration(y, face, couponRate, m, ppy)
```

```python
#FizzBuZZ
import numpy as np

def FizzBuzz(start, finish):

    nums = np.arange(start, finish + 1)

    result = nums.astype(object)   # Important!

    fizz = (nums % 3 == 0)
    buzz = (nums % 5 == 0)

    result[fizz & buzz] = "fizzbuzz"
    result[fizz & ~buzz] = "fizz"
    result[buzz & ~fizz] = "buzz"

    return result
FizzBuzz(61, 75)
```
